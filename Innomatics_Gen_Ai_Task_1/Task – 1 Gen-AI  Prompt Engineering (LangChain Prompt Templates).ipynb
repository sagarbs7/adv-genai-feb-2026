#Task 4: ChatPromptTemplate System
Objective: Use System and User roles for chat-based LLM interactions.

```python
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

def get_chat_prompt(role, topic):
    # Mapping roles to specific behaviors
    role_behaviors = {
        "teacher": "You are a helpful teacher who simplifies complex concepts.",
        "interviewer": "You are a professional interviewer assessing technical skills.",
        "motivator": "You are an encouraging mentor who inspires students."
    }

    system_template = SystemMessagePromptTemplate.from_template(role_behaviors.get(role, "You are a helpful assistant."))
    user_template = HumanMessagePromptTemplate.from_template("{topic}")

    chat_prompt = ChatPromptTemplate.from_messages([system_template, user_template])
    return chat_prompt.format_prompt(topic=topic).to_messages()

# Example Test
print(get_chat_prompt(role='teacher', topic='Neural Networks'))
```

#Task 5: Input Validation Layer
Objective: Ensure inputs match allowed criteria before processing.

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience. Default set to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone. Default set to 'formal'")
        tone = "formal"

    return audience, tone
```

```python
audience = input("Enter audience: ")
tone = input("Enter tone: ")

# Validate inputs
audience, tone = validate_inputs(audience, tone)

# Print final values
print("\nFinal Values:")
print("Audience:", audience)
print("Tone:", tone)
```

#Task 6: Prompt Generator App
Objective: A master function integrating validation and modular styles.


```python
def generate_prompt(topic, audience, tone, style):
    # 1. Validate inputs
    v_audience, v_tone = validate_inputs(audience, tone)

    # 2. Select Style Template
    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Generate interview questions on {topic} for {audience} level in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    selected_template = styles.get(style, styles["teaching"])

    # 3. Use LangChain PromptTemplate
    prompt_obj = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=selected_template
    )

    return prompt_obj.format(topic=topic, audience=v_audience, tone=v_tone)

# Example Output
print(generate_prompt("Generative-AI", "beginner", "fun", "storytelling"))
```

Objective: Run the same template object with 5 unique inputs.
