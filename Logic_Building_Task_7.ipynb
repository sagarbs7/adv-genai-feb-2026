# # Problem 1 - Smart Light Controller

```python
class Light:

    # Constructor method
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    # Turn the Light ON
    def turn_on(self):
        self.status = "ON"

    # Turn the Light OFF
    def turn_off(self):
        self.status = "OFF"

    # Display current light status
    def display_status(self):
        print(self.light_name + " is " + self.status)

# Creation of Object
light = Light("Bedroom Light")

# Turning light ON
light.turn_on()

# Displaying the current status of the light
light.display_status()
```

# Problem 2 - Employee ID Card System

```python
class employee:

    # Constructor for Initalizing the Employee Details
    def __init__(self,name,id,dept):
        self.name = name
        self.id = id
        self.dept = dept

    # Displaying
    def display_card(self):
        print("Name : ",self.name)
        print("Id : ",self.id)
        print("Department : ",self.dept)

# Creating an Object
emp1 = employee("sagar","Ew2022","CS")

# Displaying the Employee Details
emp1.display_card()
```

# Problem 3 - Mobile Contact Record

```python
# Problem 3 - Mobile Contact Record
class Contact:

    # Constructor to initialize contact name and phone number
    def __init__(self,name,phone):
        self.name = name
        self.phone = phone

    # Method to display contact information
    def display_contact(self):
        print("Name:", self.name)
        print("Phone:", self.phone)
        print("Contact Saved")

# Creating a object
contact1 = Contact("sagar", "87548983737")

# Displaying the Details
contact1.display_contact()
```

# Problem 4 - Product Price Tag Generator

```python
# Problem 4 - Product Price Tag Generator
class Product:

    # Constructor
    def __init__(self, name, price):
        self.name = name 
        self.price = price

    # Method to display formatted price tag
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

# Creating a object
product1 = Product("sanmsung", 170000)

# Displaying the price tag
product1.display_price_tag()
```

# Problem 5 - Movie Rating Display System

```python
class Movie:

    # Constructor
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    # Method to display movie information
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Creating a movie object
movie1 = Movie("troy", 4.7)

# Displaying the movie details
movie1.display_movie()
```

# Problem 6 - Delivery Address Manager

```python
class Delivery:

    # Constructor
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to display delivery details
    def display_details(self):
        print("Delivery Details :")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Creating an object with sample input
order1 = Delivery("Sagar", "Manjunathnagar near gopalan arcade")

# Displaying the delivery information
order1.display_details()
     
```
