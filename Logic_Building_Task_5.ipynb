# Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement():
    total_likes = 0
    
    n = int(input("Enter number of like entries: "))
    
    # Loop to take user input
    for i in range(n):
        like = int(input(f"Enter likes {i+1}: "))
        total_likes += like
    
    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Display output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Function Call
analyze_post_engagement()
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python

def rainfall_checker():
    
    n = int(input("Enter number of rainfall readings: "))
    total = 0
    
    for i in range(n):
        rain = float(input(f"Enter rainfall {i+1}: "))
        total += rain
    
    required_level = float(input("Enter required rainfall level: "))
    
    average = total / n
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)


# Function Call
rainfall_checker()
```

# Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts():
    
    n = int(input("Enter number of usernames: "))
    usernames = []
    
    for i in range(n):
        name = input(f"Enter username {i+1}: ")
        usernames.append(name)
    
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Function Call
detect_duplicate_accounts()
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crop_filter():
    
    n = int(input("Enter number of crop prices: "))
    prices = []
    
    for i in range(n):
        price = int(input(f"Enter price {i+1}: "))
        prices.append(price)
    
    premium_crops = []
    
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)


# Function Call
premium_crop_filter()
```

# Problem 7: System Monitoring – Application Health Checker

```python
def application_health_checker():
    
    error_count = int(input("Enter error count: "))
    
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", error_count)
    print("System Status:", status)


# Function Call
application_health_checker()
```

# Problem 9: E-Learning – Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_list):
    
    total_classes = len(attendance_list)
    attended_classes = 0
    
    # Loop to count attended days
    for day in attendance_list:
        if day == 1:   # 1 means Present, 0 means Absent
            attended_classes += 1
    
    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    
    # Conditional Check
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    # Output
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)


# Example Data (1 = Present, 0 = Absent)
attendance_data = [1,1,1,0,1,1,1,1,0,1]   # 8/10 = 80%

# Function Call
check_exam_eligibility(attendance_data)
    
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    
    bill = 0
    
    # Slab Calculation Logic
    if units <= 100:
        bill = units * 3
        
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
        
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

        # Usage Classification Logic
    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    
    # Return Multiple Values
    return bill, usage_status


# Example Usage
units_consumed = 250

total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Classification:", status)
```
